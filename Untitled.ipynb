```python
def split_tensor_into_chunks(tensor):
    # extract time tensor
    time_tensor = tensor.extract('t')
    # find chunks of same time step variable
    chunks = []
    start_idx = 0
    for i in range(1, len(time_tensor)):
        if not torch.equal(time_tensor[i], time_tensor[start_idx]): # if two time variables are not equal
            chunks.append(tensor[start_idx:i]) # create a new chunk
            start_idx = i
    chunks.append(tensor[start_idx:])
    return chunks
```

```python
import torch

from pina.problem import SpatialProblem, TimeDependentProblem
from pina.geometry import CartesianDomain
from pina.operators import grad
from pina.equation import FixedValue, Equation
from pina import Condition, Trainer, Plotter
from pina.solvers import CasualPINN
from pina.model import FeedForward


alpha = 5
beta = 0.5
gamma = 0.005
domain_x = [-1, 1]
domain_t = [0, 1]

class KuramotoSivashinsky(SpatialProblem, TimeDependentProblem):

    temporal_domain = CartesianDomain({'t' : domain_t})
    spatial_domain = CartesianDomain({'x' : domain_x})
    output_variables = ['u']

    def kuramoto_sivashinsky(input_, output_):
        grad_ = grad(output_, input_, d=['x', 't'])
        u_t = grad_.extract(['dudt'])
        u_x = grad_.extract(['dudx'])
        u_xx = grad(u_x, input_, d=['x'])
        u_xxx = grad(u_xx, input_, d=['x'])
        u_xxxx = grad(u_xxx, input_, d=['x'])
        return u_t + alpha * output_ * u_x + beta * u_xx + gamma * u_xxxx

    @staticmethod
    def initial(input_):
        return - torch.sin(torch.pi * input_.extract(['x']))
    
    def initial_condition(input_, output_):
        return output_ - KuramotoSivashinsky.initial(input_)
    
    conditions = {
        # 'gamma1': Condition(
        #     location=CartesianDomain({'x': domain_x[0], 't' : domain_t}),
        #     equation=FixedValue(0.0)),
        # 'gamma2': Condition(
        #     location=CartesianDomain({'x': domain_x[1], 't' : domain_t}),
        #     equation=FixedValue(0.0)),
        # 't0': Condition(location=CartesianDomain({'x': domain_x, 't': 0}), equation=Equation(initial_condition)),
        'D': Condition(location=CartesianDomain({'x': domain_x,  't': domain_t}), equation=Equation(kuramoto_sivashinsky)),}

class HardNet(torch.nn.Module):
    def __init__(self, model):
        super().__init__()

        self.model = model

    def forward(self, x):
        out = self.model(x) * (1. - torch.abs(x.extract(['x'])))
        return out
    
# make the problem
burgers_problem = KuramotoSivashinsky()
burgers_problem.discretise_domain(n=64, mode='grid', variables = 't', locations=['D'])
burgers_problem.discretise_domain(n=32, mode='grid', variables = 'x', locations=['D'])
# burgers_problem.discretise_domain(n=200, mode='random', locations=['gamma1', 'gamma2', 't0'])

model = HardNet(FeedForward(
    len(burgers_problem.input_variables),
    len(burgers_problem.output_variables),
    inner_size=256,
    n_layers=5))

pinn = CasualPINN(problem = burgers_problem, model=model)
trainer = Trainer(solver=pinn, max_epochs=100, accelerator='cpu', batch_size=None)
trainer.train()
pl = Plotter()
pl.plot(solver=pinn)

```
